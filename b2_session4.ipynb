```python
#STUDENT PERFORMANCE ANALYSER

"""● Read session4_dumny_students.csv.
● Create an average column using math, science, and english.
● Create a result column: Pass if average > 40, otherwise Fail.
● Print class average.
● Print highest average and lowest average subject wise.
● Print top student or students.
● Print students with attendance below 75.
● Print students who submitted fewer than 6 assignments.
● Save final output as session4_student report.csv."""
```

```python
import pandas as pd
import numpy as np

def get_result(avg):
  if avg>=40:
    return "Pass"
  return "Fail"

df = pd.read_excel("session4_dummy_students_data.csv.xlsx")

df["average"] = df[["math","science","english"]].mean(axis=1)
df["result"] = df["average"].apply(get_result)

print("Full_Data:")
print(df)

print("class average:",df["average"].mean())
print("Highest average:",df["average"].max())
print("Lowest average:",df["average"].min())

low_assignment_student = df[df["assignments_submitted"] < 6]
print("Students with fewer than 6 assignments:")
print(low_assignment_student[["name", "assignments_submitted"]])
```
